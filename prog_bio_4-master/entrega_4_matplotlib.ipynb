### Ejercicio 1

Representa en un único gráfico las funciones arcoseno y arcocoseno en el intervalo [-pi/4, pi/4] 

```python
%matplotlib inline
# Respuesta

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x = np.arange(-np.pi/4,np.pi/4, 0.1)

arcoseno_x = np.arccos(x)
arcocoseno_x = np.arcsin(x)

plt.plot(x, arcoseno_x, 'b', x, arcocoseno_x, 'r')
plt.xlabel('x')
plt.ylabel('funciòn')
plt.legend(('arcsin(x)', 'arccos(x)'))


plt.show()
```

### Ejercicio 2

Representa en un gráfico 3D la función `f(x,y) = cos(x) + 2*sen(y)` en el intervalo [-2*pi, 2*pi]

```python
%matplotlib inline

# Respuesta
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-2*np.pi,2*np.pi, 0.1)
y = np.arange(-2*np.pi,2*np.pi, 0.1)

[X,Y] = np.meshgrid(x,y)

F = (np.cos(X) + np.sin(Y)*2)

figura = plt.figure()

ax = figura.add_subplot(1, 1, 1, projection='3d')

ax.plot_surface(X, Y, F, cmap='coolwarm')

plt.show()
```

### Ejercicio 4

Dados los siguientes datos de meteorología registrados en el aeropuerto de Barcelona, representa en uno o varios gráficos los valores de temperatura media anual, máxima, mínima, media de lluvias mensual y humedad relativa.

Las posiciones del vector `meteo` son las siguientes:

* Nombre mes
* Temperatura media anual del mes (°C)
* Temperatura máxima media (°C)
* Temperatura mínima media (°C)
* Media de lluvias mensual (mm)
* Humedad media relativa (%)


```python
%matplotlib inline

meteo = [["January", 9.2, 13.6, 4.7, 37, 70],
["February",9.9, 14.3, 5.4, 35, 70],
["March", 11.8, 16.1, 7.4, 36, 70],
["April", 13.7, 18.0, 9.4, 40, 69],
["May", 16.9, 21.1, 12.8, 47, 70],
["June", 20.9, 24.9, 16.8, 30, 68],
["July", 23.9, 28.0, 19.8, 21, 67],
["August", 24.4, 28.5, 20.2, 62, 68],
["September", 21.7, 26.0, 17.4, 81, 70],
["October", 17.8, 22.1, 13.5, 91, 73],
["November", 13.0, 17.3, 8.6, 59, 71],
["December", 10.0, 14.3, 5.7, 40, 69]]


# Respuesta

# Respuesta
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

array_meteo = np.array(meteo, dtype=object)
array_meteo_trans = np.transpose(array_meteo)

months = array_meteo_trans[0]
temp_avg = array_meteo_trans[1]
temp_max = array_meteo_trans[2]
temp_min = array_meteo_trans[3]
rain_avg = array_meteo_trans[4]
humidity_avg = array_meteo_trans[5]

#Grafico para temperaturas
plt.plot(months,temp_avg,'g')
plt.plot(months,temp_max,'r')
plt.plot(months,temp_min,'b')
plt.xlabel('Months')
plt.ylabel('Temprature (ºC)')

plt.legend(['AVG_Temp', 'Max_Temp', 'Min_Temp'])
plt.xticks(rotation=45)

plt.show()

#Grafico de lluvia
plt.xticks(rotation=45);
plt.xlabel('Months')
plt.ylabel('AVG_Rain')
plt.bar(months,rain_avg);
plt.show()

# Gráfico de humedad
plt.plot(months,humidity_avg)
plt.xlabel('Months')
plt.ylabel('AVG_Humidity')
plt.xticks(rotation=45);
plt.show()
```
